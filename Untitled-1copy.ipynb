```python
from math import exp, sqrt, cos, sin
import itertools

import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
from scipy.optimize import fsolve
```

```python
hbar_w_op = 0.2
hbar = 6.582119569*10**(-16)
w_ph = hbar_w_op / hbar
G_op2 = 0.5 * 0.5
eta = 7.85 * 10**(-3)
a = 0.142 * 10**(-9)
t = 2.52
kb = 8.6173303 * 10**(-5)
pi = np.pi
Ef = 0.5
kF=Ef/3*2/a/t
```

```python
print(kF)
```

```python
def fermi_dirac(E, Ef, T): 
    if T > 0:
       if E > Ef:
           return exp((-E + Ef) / (kb * T)) / (exp((-E + Ef) / (kb * T)) + 1)
       else:
           return 1 / (exp((E - Ef) / (kb * T)) + 1)
    else:
        if E < Ef: 
            return 0 
        else: 
            return 1

def bose_einstein(E,T): 
    assert T > 0
    return 1 / (exp(E / (kb * T)) - 1)

def chemical_potential(T):
    return Ef - pi**2 / 6 / Ef * (kb * T)**2

def eps(k):
    return 3 * a * t / 2 * k

def ss_sum_part(k, w, w_phonon, T, u):
    def summand(s1, s2):
        return ((bose_einstein(hbar * w_phonon, T) + fermi_dirac(s1 * s2 * eps(k), u, T)) / (hbar * w + s1 * eps(kF) - s1 * eps(k) + s2 * hbar * w_phonon + eta * 1j))
    sum = 0
    for s1, s2 in itertools.product([-1,1], [-1, 1]):
        sum = sum + summand(s1, s2)
    return sum.real



def mem_fn(w, w_phonon, u=Ef, T=50):
    def find_max():
        max_point, kmax = 0, 0
        min_point_1, kmin1 = 0, 0
        min_point_2, kmin2 = 0, 0
        for k in np.arange(0, kF, 10**5):
            ss = ss_sum_part(k, w, w_ph, T, u)
            if max_point < ss:
                max_point = ss
                kmax = k
            if min_point_2 > ss:
                if ss < min_point_1:
                    min_point_2, kmin2 = min_point_1, kmin1
                    min_point_1, kmin1 = ss, k
                else:
                    min_point_2, kmin2 = ss, k
        return [kmax, kmin1, kmin2]
    print(find_max())
    
    integral_k = integrate.quad(ss_sum_part, 0, kF, args=(w, w_phonon, T, u), limit=80000, points=find_max())[0]
    return -G_op2 * a * integral_k / hbar / (2*pi)
```

```python
ws = np.arange(0, 0.5, 0.1)
T=50
ks = np.arange(0, kF, 10**6)
for w in ws:
    ss_sums = []
    for k in ks:
        ss = ss_sum_part(k, w, w_ph, T, chemical_potential(T))
        ss_sums.append(ss)
    plt.plot(ks, ss_sums, label = f"{w} eV")
plt.legend()
```

```python
result_real = []
sums = []
ws = np.arange(0, 1, 0.01)
T=50
for w in ws:
    mem = mem_fn(w=w, w_phonon=w_ph, u=chemical_potential(T))
    result_real.append(mem.real)
    print(mem)
plt.plot(ws, result_real)
```
