```python
from math import exp, sqrt, cos, sin
import itertools

import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
from scipy.optimize import fsolve
```

```python
hbar_w_op = 0.2
hbar = 6.582119569*10**(-16)
w_ph = hbar_w_op / hbar
G_op2 = 0.5 * 0.5
eta = 7.85 * 10**(-3)
a = 0.142 * 10**(-9)
t = 2.52
kb = 8.6173303 * 10**(-5)
pi = np.pi
kf = 1.799 * 10**10
vf = 3 * a * t / 2 / hbar
Ef = 0.5
```

```python
def fermi_dirac(E, Ef, T): 
    if T > 0:
       if E > Ef:
           return exp((-E + Ef) / (kb * T)) / (exp((-E + Ef) / (kb * T)) + 1)
       else:
           return 1 / (exp((E - Ef) / (kb * T)) + 1)
    else:
        if E < Ef: 
            return 0 
        else: 
            return 1

def bose_einstein(E,T): 
    assert T > 0
    #return exp((+E)/(kb*T))/(exp((+E)/(kb*T))-1)
    return 1 / (exp(E / (kb * T)) - 1)

def eps(k):
    return 3 * a * t / 2 * k

def ss_sum_part(k, w, w_phonon, T, u):
    def summand(s1, s2):
        return (bose_einstein(hbar * w_phonon, T) + fermi_dirac(s1 * s2 * eps(k), u, T)) * (hbar * w + s1 * eps(kf) - s1 * eps(k) + s2 * hbar * w_phonon) / ((hbar * w + s1 * eps(kf) - s1 * eps(k) + s2 * hbar * w_phonon)**2 + eta**2)
    sum = 0
    for s1, s2 in itertools.product([-1,1], [-1, 1]):
        sum = sum + summand(s1, s2)
    return sum * k 

def mem_fn(w, w_phonon, u=Ef, T=50):
    integral_k = integrate.quad(ss_sum_part, 0, 10**10, args=(w, w_phonon, T, u), limit=80000)[0]
    return -G_op2 * a * integral_k / hbar / (2*pi)


def particle_density(E, u, T):
    g = E / (2 * pi * hbar**2 * vf**2)
    fs = -fermi_dirac(-E, -u, T) + fermi_dirac(-E, u, T)
    return g * fs

def particle_density_int(u, kf, T):
    na = kf**2 / 4 / pi
    #na=1/3/pi**2*Kf**3/(vf*hbar)**3
    return na - integrate.quad(particle_density, 0, 1, args=(u,T), limit=8000)[0]
```

```python
ks = np.arange(0, 1000000, 0.1)
sums = []
for k in ks:
    sums.append(ss_sum_part(k, w=0.5, w_phonon=w_ph, T=50, u=0.5))
plt.plot(ks, sums)
```
